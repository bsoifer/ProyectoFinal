```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

# Leer csv
df_tiendas_m1 = pd.read_csv('/content/drive/MyDrive/2024 2Q/Proyecto Final/TP Proyecto/Entrega 3/Codigo E3/tiendas_caba_v7.csv')
df_tiendas_m2 = pd.read_csv('/content/drive/MyDrive/2024 2Q/Proyecto Final/TP Proyecto/Entrega 3/Codigo E3/tiendas_caba_m2.csv')
```

```python
# Contar las filas con al menos un NaN
num_filas_con_nan = df_tiendas_m2.isna().any(axis=1).sum()

print(f"Número de filas con al menos un NaN: {num_filas_con_nan}")
```

```python
df_tiendas_m2.shape
```

```python
# merge
df_tiendas_m2 = df_tiendas_m2.merge(df_tiendas_m1[['id', 'cluster']],
                                    left_on='customer_id',
                                    right_on='id',
                                    how='left')

df_tiendas_m2 = df_tiendas_m2.drop(columns='id')

print(df_tiendas_m2.head())
```

```python
df_tiendas_m2 = df_tiendas_m2.dropna()
```

```python
df_tiendas_m2.columns
```

```python
from sklearn.model_selection import train_test_split

# Realizar el train-test split con estratificación directamente sobre el DataFrame
train, test = train_test_split(
    df_tiendas_m2,
    test_size=0.2,
    random_state=42,
    stratify=df_tiendas_m2['cluster']
)
```

```python
train.to_csv('tiendas_caba_m2_train.csv', index=False)
test.to_csv('tiendas_caba_m2_test.csv', index=False)

# Descargar el archivo CSV en tu computadora
from google.colab import files
files.download('tiendas_caba_m2_train.csv')
files.download('tiendas_caba_m2_test.csv')
```
