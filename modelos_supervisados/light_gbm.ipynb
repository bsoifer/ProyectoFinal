```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

# Leer csv
df_tiendas_train = pd.read_csv('/content/drive/MyDrive/2024/2do Cuatri - 2024/Proyecto Final/TP Proyecto/Entrega 3/Codigo E3/tiendas_caba_m2_train.csv')
df_tiendas_val = pd.read_csv('/content/drive/MyDrive/2024/2do Cuatri - 2024/Proyecto Final/TP Proyecto/Entrega 3/Codigo E3/tiendas_caba_m2_test.csv')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

# Definir las características (X) y la variable objetivo (y)
x_train = df_tiendas_train.drop(columns=['customer_id', 'cluster'])  # Eliminamos customer_id y cluster
y_train = df_tiendas_train['cluster']

# Definir las características (X) y la variable objetivo (y)
x_val = df_tiendas_val.drop(columns=['customer_id', 'cluster'])  # Eliminamos customer_id y cluster
y_val = df_tiendas_val['cluster']
```

```python
x = pd.concat([x_train, x_val], axis=0)
y = pd.concat([y_train, y_val], axis=0)
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, classification_report

# Definir el modelo
model = LGBMClassifier(
    random_state=42,
    colsample_bytree=1.0,
    learning_rate=0.01,
    max_depth=30,
    min_child_samples=20,
    min_split_gain=0,
    n_estimators=1000,
    subsample=0.8)

# Entrenar el modelo
model.fit(x_train, y_train)

# RESULTADOS VAL
print("RESULTADOS VAL:")
y_pred_val = model.predict(x_val)
print(f"Accuracy: {accuracy_score(y_val, y_pred_val)}")
print(classification_report(y_val, y_pred_val))

# RESULTADOS TRAIN
print("RESULTADOS TRAIN:")
y_pred_train = model.predict(x_train)
print(f"Accuracy: {accuracy_score(y_train, y_pred_train)}")
print(classification_report(y_train, y_pred_train))

# RESULTADOS TODO
print("RESULTADOS TODO:")
y_pred_all = model.predict(x)
print(f"Accuracy: {accuracy_score(y, y_pred_all)}")
print(classification_report(y, y_pred_all))
```

```python
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, classification_report

# Definir el modelo base
model = LGBMClassifier(random_state=42)

# Definir los hiperparámetros para el grid search
param_grid = {
    'n_estimators': [100, 500, 1000],  # Número de árboles
    'max_depth': [10, 20, 30],         # Máxima profundidad
    'min_child_samples': [10, 20, 30], # Equivalente a min_samples_leaf
    'learning_rate': [0.01, 0.1, 0.3], # Tasa de aprendizaje
    'subsample': [0.8, 1.0],           # Fracción de muestras usadas para entrenar cada árbol
    'colsample_bytree': [0.8, 1.0],    # Fracción de características usadas por árbol
    'min_split_gain': [0, 10, 50]      # Equivalente a min_samples_split
}

# Definir el grid search con validación cruzada
grid_search = GridSearchCV(estimator=model,
                           param_grid=param_grid,
                           scoring='accuracy',  # Métrica a optimizar
                           cv=5,                # Número de folds en validación cruzada
                           verbose=1,           # Mostrar progreso
                           n_jobs=-1)           # Usar todos los núcleos disponibles

# Ejecutar el grid search
grid_search.fit(x_train, y_train)

# Imprimir los mejores parámetros encontrados
print("Mejores hiperparámetros encontrados:")
print(grid_search.best_params_)
```
