```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que df_tiendas es tu DataFrame y tiene las columnas adecuadas
# Lista de columnas que indican si la tienda está abierta en un día específico
dias = ['mon_open', 'tue_open', 'wed_open', 'thu_open', 'fri_open', 'sat_open', 'sun_open']

# Contar el número de tiendas abiertas cada día
cantidad_abiertos = {dia: df_tiendas[dia].sum() for dia in dias}

# Convertir el diccionario a DataFrame para facilitar el gráfico
df_abiertos = pd.DataFrame(list(cantidad_abiertos.items()), columns=['Día', 'Cantidad de Tiendas Abiertas'])

# Crear el gráfico de barras con color lightblue y bordes
plt.figure(figsize=(10, 6))
plt.bar(df_abiertos['Día'], df_abiertos['Cantidad de Tiendas Abiertas'], color='lightblue', edgecolor='black')
plt.title('Cantidad de Tiendas Abiertas por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Cantidad de Tiendas Abiertas')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
from scipy import stats

# Contar el número de tiendas abiertas por día de la semana
counts = {
    'Lunes': df_tiendas['mon_open'].sum(),
    'Martes': df_tiendas['tue_open'].sum(),
    'Miércoles': df_tiendas['wed_open'].sum(),
    'Jueves': df_tiendas['thu_open'].sum(),
    'Viernes': df_tiendas['fri_open'].sum(),
    'Sábado': df_tiendas['sat_open'].sum(),
    'Domingo': df_tiendas['sun_open'].sum()
}

# Convertir los datos a listas para cada día
lunes = df_tiendas['mon_open'].tolist()
martes = df_tiendas['tue_open'].tolist()
miercoles = df_tiendas['wed_open'].tolist()
jueves = df_tiendas['thu_open'].tolist()
viernes = df_tiendas['fri_open'].tolist()
sabado = df_tiendas['sat_open'].tolist()
domingo = df_tiendas['sun_open'].tolist()

# Realizar el ANOVA
f_stat, p_value = stats.f_oneway(lunes, martes, miercoles, jueves, viernes, sabado, domingo)

print(f'Estadístico F: {f_stat:.4f}')
print(f'Valor p: {p_value:.4f}')
```

```python
import matplotlib.pyplot as plt

# Contar la cantidad de supermercados chinos y no chinos
cantidad_supermercados_chinos = df_tiendas['supermercado_chino'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(cantidad_supermercados_chinos.index.map({True: 'Chinos', False: 'No Chinos'}),
        cantidad_supermercados_chinos.values,
        color='lightblue', edgecolor='black')

plt.title('Cantidad de Tiendas - Supermercados Chinos vs No Chinos')
plt.xlabel('Tipo de Tienda')
plt.ylabel('Cantidad de Tiendas')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Contar la cantidad de tiendas que tienen y no tienen bicicleteros
cantidad_bicicleteros = df_tiendas['hay_bicicleteros_200m'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(cantidad_bicicleteros.index.map({1: 'Con Bicicleteros', 0: 'Sin Bicicleteros'}),
        cantidad_bicicleteros.values,
        color='lightblue', edgecolor='black')

plt.title('Cantidad de Tiendas - Con y Sin Bicicleteros en un Radio de 200m')
plt.xlabel('Tipo de Tienda')
plt.ylabel('Cantidad de Tiendas')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Unir todas las columnas de categorías en un solo DataFrame
categorias_df = df_tiendas[['categoria_1', 'categoria_2', 'categoria_3', 'categoria_4', 'categoria_5']]

# Crear una lista de categorías únicas para cada tienda
def contar_categorias_unicas(row):
    return len(set(row.dropna()))  # Eliminar valores NaN y contar categorías únicas

# Aplicar la función a cada fila para contar las categorías únicas por tienda
categorias_unicas_count = categorias_df.apply(contar_categorias_unicas, axis=1)

# Graficar la diversidad de categorías por tienda
plt.figure(figsize=(10, 6))
plt.hist(categorias_unicas_count, bins=range(1, 7), color='lightblue', edgecolor='black', align='left')
plt.title('Diversidad de Categorías Compradas por Tienda')
plt.xlabel('Número de Categorías Únicas')
plt.ylabel('Número de Tiendas')
plt.xticks(range(1, 6))
plt.show()
```
